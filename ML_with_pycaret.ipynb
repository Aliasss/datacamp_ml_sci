```python
!pip install pycaret
from pycaret.datasets import get_data
# df = get_data('credit')
df = get_data('juice')
```

```python
print(df.shape)
display(df.head())
train = df.sample(frac=0.95, random_state=49)
test = df.drop(train.index)
train.reset_index(inplace=True, drop=True)
test.reset_index(inplace=True, drop=True)
print(f"train shape: {train.shape}, test shape: {test.shape}")
```

```python
# !pip install -U imbalanced-learn
!pip install delayed
```

```python
# import imblearn
from pycaret.classification import *
# 분류 패키지의 모든 모듈을 임포트한다.

exp = setup(data = train,  
            target = 'Purchase',
            
            normalize = True, 
            normalize_method = 'minmax', # 기본은 zscore 
            
            transformation = True, 
            
            fold = 5, # 기본적으로 10 fold 로 training 한다.
            fold_shuffle=True,
            
            ignore_features = ['PriceDiff'], # 제외할 컬럼
            numeric_features = ['PriceCH','PriceMM'],
            categorical_features = ['Store7'], # 지정하면 onehotencoding된다. 
            date_features = [], # 날짜 feature를 년월일시 로 바꿔서 onehotencoding 해준다. 
            
            silent = True, # setup 시 중간에 피쳐속성 확인하고 엔터 쳐줘야하는데 알아서 넘어가게 해준다.
           
            session_id = 49, # random state number 지정
            
            use_gpu = False, # gpu 사용 옵션
            
            feature_selection = True,
            feature_selection_method = 'classic', # or 'boruta' 
            # classic 은 permutation importance 기반.
            
            fix_imbalance = True, # data imbalance 를 sampling method로 보정 (dafault = SMOTE)
            # fix_imbalance_method = imblearn.OverSampling.RandomOverSampler()
            
            # custom_pipeline = pipe, preprocess =False 
            # 두 개는 세트, 사용자가 원하는 파이프라인을 구성할 수 있다. 
            )
```

```python
"""
compare_models(): 다양한 모델 학습 후 성능 비교
fold: cross_validation의 fold를 지정 (default = 10)
sort: 정렬기준 지표 설정
n_select: 상위 n개의 모델 결과만 출력
"""

# 특정 모델만 비교
# best1 = compare_models(include = ['lr','lightgbm','rf'], sort='F1')
# best2 = compare_models(exclude = ['XGboost'], sort='F1')
# best3 = compare_models(sort='Accuracy', n_select = 3, budget_time = 0.5)

# 모든 모델 실행 및 비교
# compare_models()
best_models = compare_models(sort='f1', n_select=3, fold=5)
```

```python
"""
create_model(): 하나의 모델 적합
fold: cross_validation의 fold 지정 (default = 10)
"""
xgb = create_model('xgboost', fold=5)
```

```python
"""
tune_model(model): 모델의 하이퍼파라미터 튜닝
optimize: 평가 metric 지정
"""

tuned_xgb = tune_model(xgb, fold=5, optimize='f1')
```

```python
tuned_xgb
```

```python
evaluate_model(tuned_xgb)
```

```python
"""
finalize_model(): 최종 모델로 설정 후 마지막 학습 진행
predict_model(): 예측 결과를 'Label' 변수에 저장
"""
final_model = finalize_model(tuned_xgb)
prediction = predict_model(estimator=final_model, data = test)
prediction
```
